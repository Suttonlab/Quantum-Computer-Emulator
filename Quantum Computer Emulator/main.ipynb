**GROVER'S**  

```python
#To run Grover's algorithm, define the number of qubit's to use in a Quantum Circuit (qubit_count), and the state to search for (key_value). 
#The function "grover" will return the end amplitudes of the circuit.
qubit_count = 6
span = (1<<qubit_count)
key_value = int(span/2)
end_values = grover(qubit_count,key_value,track=False)  
```

```python
#The same as above, except track values for visualisation.
end_values_t, tracked_values = grover(qubit_count,key_value,track=True)  
plots.gplot(end_values_t,span,key_value)
```

```python
#Apply Grover's to a sentence. It will prompt an input for the sentence, and then for the word searched for.
circuit_end_w, track, key_value_w, span_w = grover(track=True,wordsearch=True)
plots.gplot(circuit_end_w,span_w,key_value_w)
```

```python
#Try a sentence with repeating words
circuit_end_w, track, key_value_w, span_w = grover(qubit_count,track=True,wordsearch=True)
plots.gplot(circuit_end_w,span_w,key_value_w)
```
